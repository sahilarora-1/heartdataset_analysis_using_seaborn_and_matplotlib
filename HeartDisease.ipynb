```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("C:\\datasets\\heart_cleveland_upload.csv")
```

```python
dataframe=pd.DataFrame(data)
```

```python
#number of people vs age
```

```python
age=dataframe['age'].value_counts()
plt.scatter(age.index,age.values,color="red",marker='o')
plt.grid(linestyle=":",color="black")
plt.title("Distribution of age")
plt.xlabel("Age")
plt.ylabel("Number of people")
plt.savefig("distribution_of_age_1.png")
plt.show()
```

```python
chest_pain=dataframe['cp'].value_counts().sort_index()
plt.bar(['typical angina','atypical angina','non-anginal pain','asymptomatic'], chest_pain.values, color=["red","pink","purple","blue"])
plt.title("Types of Chest Pain")
plt.xlabel("type of pains")
plt.ylabel("number of people with pain")
plt.savefig("typesofchestpain_1.png")
plt.show()
```

```python
fbs=dataframe['fbs'].value_counts()
plt.bar(['High','Normal'],fbs.values,color=['red','blue'])
plt.title('Blood Sugar')
plt.xlabel('High OR Normal')
plt.ylabel('Number of People')
plt.savefig("barchart_showing_sugarlevel3.png")
plt.show()
```

```python
heartrate=dataframe['thalach'].value_counts().sort_index()
heartrate
```

```python
plt.hist(dataframe['thalach'],color='darkgreen',edgecolor='black',label='heart_rate')
plt.xlabel('number of people')
plt.ylabel('Heartbeat Rate Value')
plt.title('Heartbeat Rate')
plt.savefig("histchartshowing_heartbeatrate.png")
plt.show()
```

```python
vessels=dataframe['ca'].value_counts()
vessels
```

```python
plt.pie(vessels.values,labels=['Zero','One','Two','Three'],autopct='%1.1f%%')
plt.title('Number of vessels visible')
plt.savefig("numberofvessselsvisible_5.png")
plt.show()
```

```python
## USING SEABORN
import seaborn as sns
dataframe['sex'] = dataframe['sex'].replace({0: "Female", 1: "Male"})
```

```python
dataframe
```

```python
sns.boxplot(data=dataframe, x="sex", y="age", hue="condition", palette="Set1", dodge=True)
sns.swarmplot(data=dataframe, x="sex", y="age", hue="condition", dodge=True, palette="Set1", alpha=0.5)
plt.title("Age Distribution by Sex and Heart Condition")
plt.ylabel("Age")
plt.xlabel("Sex")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.savefig("agedistributionbysex_6.png")
plt.show()
```

```python
dataframe['thal'] = dataframe['thal'].replace({
    0: "Normal",
    1: "Fixed def",
    2: "Reversible def"
})
```

```python
dataframe['thal'].value_counts()
```

```python
# The plot shows how heart disease cases vary by Thallium stress test results.  
# Patients with a 'Reversible Defect' are more likely to have heart disease compared to other groups.  

sns.countplot(data=dataframe,x="thal",hue="sex",palette=["orange", "lightblue"], hue_order=["Male", "Female"])
plt.title('Thallium stress test vs Condition')
plt.savefig("thalliumstress_vs_condition_7.png")
```

```python
#cholestrolincrease with high blood pressure
sns.regplot(data=dataframe, x="trestbps", y="chol", scatter_kws={"alpha":0.5}, line_kws={"color":"red"})
plt.title("Resting BP vs Cholesterol")
sns.set_context('paper')
sns.despine()
plt.savefig("BP_vs_cholestrol.png")
plt.show()
```

```python
sns.lineplot(data=dataframe, x="age", y="thalach", hue="condition", estimator="mean")
plt.title("maximum heart rate by Age")
plt.savefig("Heart_ratebyage_9.png")
plt.show()
```

```python
sns.kdeplot(data=dataframe, x="age", hue="condition", fill=True, common_norm=False, palette="Set2", alpha=0.5)
plt.title("Age Distribution by Heart Disease Condition")
plt.xlabel("Age")
plt.ylabel("Density")
plt.savefig("age_by_disease_10.png")
plt.show()
```

```python
sns.histplot(dataframe['age'], bins=20, kde=True, color="red")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.savefig("agedistribution_using_hist_11.png")
plt.show()
```
